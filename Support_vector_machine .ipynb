**Importing Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Importing The Dataset**

```python
dataset = pd.read_csv('/content/drive/MyDrive/Social Network Advertisement/Social_Network_Ads.csv');
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

print(X[:10])
print()

print(y[:10])
print()
```

**Splitting the dataset into the Training set and test**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train[:5])
print()

print(X_test[:5])
print()

print(y_train[:5])
print()

print(y_test[:5])
print()
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
print(X_train[:5])
print()

print(X_test[:5])
print()
```

**Training The Support Vector Machine Model On The Training set**

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```

**Predicting A New Result**

```python
print(classifier.predict(sc.transform([[30, 87000]])))
```

**Predicting The Test Set Results**

```python
y_pred = classifier.predict(X_test)
actual_vs_Predicted = np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test), 1)),1)
actual_vs_Predicted[:5]
```

**Making The Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print()
accuracy_of_model = accuracy_score(y_test, y_pred)
print(accuracy_of_model)
```

**Visualising The Training Set Result**

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('SVM Kernal (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

**Visualising The Test Set Result**

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('SVM Kernal (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
